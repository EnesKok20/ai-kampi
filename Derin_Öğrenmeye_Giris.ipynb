```python
import torch
import torch.nn as nn
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

housing = fetch_california_housing(as_frame=True)
df = housing.frame

X = df.drop(columns=['MedHouseVal'])
Y = df['MedHouseVal']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
y_train_tensor = torch.tensor(Y_train.values, dtype=torch.float32).reshape(-1, 1)

X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
y_test_tensor = torch.tensor(Y_test.values, dtype=torch.float32).reshape(-1, 1)
```

```python
model = nn.Sequential(
    nn.Linear(8, 16),
    nn.ReLU(),
    nn.Linear(16, 8),
    nn.ReLU(),
    nn.Linear(8, 1)
)
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 1000
for epoch in range(epochs):
    y_pred = model(X_train_tensor)
    loss = loss_fn(y_pred, Y_train_tensor)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if(epoch+1) % 10 == 0:
        print(f'Epoch: {epoch+1}/{epochs}, Loss: {loss.item():.4f}')
```

```python
model.eval()
with torch.no_grad():
    y_pred_test = model(X_test_tensor)
    test_loss = loss_fn(y_pred_test, y_test_tensor)

print(f"Test Loss (MSE): {test_loss.item():.4f}")
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

y_pred_numpy = y_pred_test.numpy()
y_test_numpy = y_test_tensor.numpy()

mae_nn = mean_absolute_error(y_test_numpy, y_pred_numpy)
r2_nn = r2_score(y_test_numpy, y_pred_numpy)

print(f"Sinir Ağı - MAE: {mae_nn:.3f}")
print(f"Sinir Ağı - R²: {r2_nn:.3f}")
```
